```python
import time
from datetime import datetime

# Utility function to format timestamp
def log(message):
    timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    print(f"[{timestamp}] {message}")

# Stage 1: Threat Modeling
def threat_modeling():
    time.sleep(1)
    risks = 4
    log(f"[Stage 1] Threat modeling completed ({risks} risks identified).")

# Stage 2: Secure Boot
def secure_boot():
    time.sleep(1)
    log("[Stage 2] Secure boot verified and system integrity validated.")

# Stage 3: Secure Key Injection
def key_injection():
    time.sleep(1)
    log("[Stage 3] Keys injected securely into the device.")

# Stage 4: OTA Firmware Update Check
def ota_update_check():
    time.sleep(1)
    update_found = True
    if update_found:
        log("[Stage 4] OTA update available and verified successfully.")
    else:
        log("[Stage 4] No OTA updates found.")

# Stage 5: Secure Decommissioning
def secure_decommission():
    time.sleep(1)
    log("[Stage 5] Device decommissioned and all secrets wiped securely.")

# Main lifecycle simulator
def run_iot_lifecycle():
    log("=== IoT Device Security Lifecycle Simulation Started ===")

    threat_modeling()
    secure_boot()
    key_injection()
    ota_update_check()
    secure_decommission()

    log("=== Simulation Completed ===")

# Run program
run_iot_lifecycle()
```
